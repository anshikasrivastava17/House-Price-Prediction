```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("simple_house.csv")
df
```

```python
sns.scatterplot(x=df['area'], y=df['price'], color='green', alpha=0.6,marker='+')
plt.title("Price vs area")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
reg = LinearRegression()

# Fit the model to the data
reg.fit(df[['area']], df['price'])

# Predict price for area = 3300
predicted_price_3300 = reg.predict([[3300]])

print("Predicted price for area 3300:", predicted_price_3300[0])


print("Coefficient (slope):", reg.coef_[0])
print("Intercept:", reg.intercept_)
```

```python
area_df = pd.read_csv("areas.csv")
area_df.head(3)
```

```python
predicted_prices = reg.predict(area_df[['area']])
```

```python
area_df['predicted_price'] = predicted_prices #added predictions to dataframe
```

```python
area_df.to_csv("predicted_prices.csv")
print(area_df)
```
