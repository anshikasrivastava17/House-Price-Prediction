```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
print("First 5 rows of dataset:\n", df.head())

if 'Id' in df.columns:
    df.drop(columns=['Id'], inplace=True)

print("\nDataset Info:")
print(df.info())
```

```python
# 1. Count Plot: Distribution of Species
plt.figure(figsize=(8, 4))
sns.countplot(x='Species', data=df)
plt.title("Class Distribution")
plt.show()
```

```python
# 2. Pairplot: Relationships between features with KDE on the diagonals
sns.pairplot(df, hue="Species", diag_kind="kde")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
# 3. Heatmap: Feature Correlation (only numerical features)
plt.figure(figsize=(8, 6))
sns.heatmap(df.iloc[:, :-1].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# 4. Line Plot: Sepal Length over Sample Index
plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x=df.index, y='SepalLengthCm', marker="o")
plt.title("Line Plot: Sepal Length over Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
# 5. Bar Plot: Count of Each Species
plt.figure(figsize=(8, 4))
species_count = df['Species'].value_counts()
sns.barplot(x=species_count.index, y=species_count.values, palette="viridis")
plt.title("Bar Plot: Count of Each Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
# 6. Scatter Plot: Sepal Length vs. Petal Length
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', hue='Species', style='Species', palette="deep")
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# 7. Box Plot: Sepal Width by Species
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Species', y='SepalWidthCm', palette="Set2")
plt.title("Box Plot: Species vs Sepal Width")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
# 8. Histogram with KDE: Petal Width Distribution with Density
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='PetalWidthCm', kde=True, stat="density", color="skyblue", edgecolor="black")
plt.title("Histogram with KDE: Petal Width Distribution")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Density")
plt.show()
```
